```python
# ===============================
# XTTS v2 – CPU ONLY
# ===============================

# 1) System dependencies
!apt-get update -qq
!apt-get install -y ffmpeg espeak-ng -qq

# 2) Python deps (CPU, STABLE versions)
!pip uninstall -y TTS torch torchvision torchaudio -q
!pip install -q torch torchaudio torchvision
!pip install -q coqui-tts
!pip install -q torchcodec

# 3) Verify TTS install
import TTS
print("TTS version:", TTS.__version__)

# 4) Convert WMV -> WAV
!ffmpeg -y -i /content/startvoice.wmv -ac 1 -ar 22050 /content/speaker.wav

import os
assert os.path.exists("/content/speaker.wav"), "speaker.wav not found"

# 5) Load XTTS (CPU)
from TTS.api import TTS as XTTS

tts = XTTS(
    model_name="tts_models/multilingual/multi-dataset/xtts_v2",
    gpu=False
)

# 6) Generate cloned speech (NOTE: This is a long text, quality might vary)
long_text = "Is ChatGPT still one of the best technology AIs?\nYes. It operates in a way somewhat similar to the human brain and does not lose track of the main prompt. "

tts.tts_to_file(
    text=long_text,
    speaker_wav="/content/speaker.wav",
    language="en",
    file_path="/content/cloned_en_long.wav"
)

print("✅ DONE: cloned_en_long.wav created")

# 7) Download result
from google.colab import files
files.download("/content/cloned_en_long.wav")
```
